```python
# EXTRACTING REGION OF INTEREST

import cv2
import glob
from google.colab.patches import cv2_imshow

# PATH TO ORIGINAL IMAGES TO BE EXTRACTED
path = glob.glob("/content/gdrive/MyDrive/CA_3/images/*.jpg") 

# LOOPING THROUGH THE FILES IN THE FOLDER CONTAINING THE IMAGES TO BE EXTRACTED
img_number = 1
for file in path:
  print(file)
  img = cv2.imread(file)

  # CUSTOMIZING THE OPTIONS OF THE IMAGES
  custom_config = r'--oem 3 --psm 11'

  # SHAPE OF THE EXTRACTED IMAGES
  img = img[230:img.shape[0]-3000,5630:img.shape[1]-290]

  # SAVING THE EXTRACTED IMAGES INTO A SEPERATE FOLDER NAMED 'Extracted_images'
  cv2.imwrite('/content/gdrive/MyDrive/CA_3/Extracted_images/image'+str(img_number)+'.jpg', img)
  img_number +=1 

  # APPLYING RECTANGULAR BOUNDING BOXES ON THE ROI
  h, w, c = img.shape
  boxes = pytesseract.image_to_boxes(img)
  for b in boxes.splitlines():
    b = b.split(' ')
    img = cv2.rectangle(img, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (255, 0, 0), 2) 
    
  # DISPLAYING THE IMAGES
  #cv2_imshow(img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
```

```python

# APPLYING GRAYSCALE AND BINARISATION TECHNIQUE ON THE EXTRACTED REGION OF INTEREST

from google.colab.patches import cv2_imshow

# PATH TO THE EXTRACTED ROI
path = glob.glob("/content/gdrive/MyDrive/CA_3/Extracted_images/*.jpg")

# LOOPING THROUGH FILES IN THE EXTRACTED ROI FOLDER
for file in path:
  print(file)
  img = cv2.imread(file) # READING THE FILES

  # GRAYSCALE FUNCTION
  grayscale_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
  img = cv2.adaptiveThreshold(grayscale_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

  # SAVING THE GRAYSCALE IMAGES INSIDE A SEPERATE FOLDER NAMED 'Grayscale_images'
  cv2.imwrite('/content/gdrive/MyDrive/CA_3/Grayscale_images/images'+str(img_number)+'.jpg', img)
  img_number +=1

  # APPLYING RECTANGULAR BOUNDING BOXES ON THE EXTRACTED GRAYSCALE IMAGES
  h, w = img.shape
  boxes = pytesseract.image_to_boxes(img)
  for b in boxes.splitlines():
    b = b.split(' ')
    img = cv2.rectangle(img, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (255, 0, 0), 2) 
  
  # IMAGE VISUALIZATION
  #cv2_imshow(img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()
```

```python
# CONTOUR DETECTION OF THE ROI IMAGES

from PIL import Image
import pytesseract
import cv2
import glob
from google.colab.patches import cv2_imshow

# PATH TO THE FOLDER CONTAINING THE ROI IMAGES
path = glob.glob("/content/gdrive/MyDrive/CA_3/Extracted_images/*.jpg")

# LOOPING THROUGH THE IMAGES IN THE ROI FOLDER
for file in path:
  print(file)
  img = cv2.imread(file)
  
  # FUNCTION FOR CONTOUR DETECTION ON THE ROI IMAGES
  img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # CONVERTING THE IMAGE COLOUR TO GRAY  
  contours = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) # FINDING CONTOURS
  cv2.drawContours(img,[],-1,(255,255,255),-1) # DRAWING THE CONTOURS
  ret, img = cv2.threshold(img, 150, 255, cv2.THRESH_BINARY) # Applying binary thresholding

  # SAVING THE EROSION IMAGES INTO A SEPERATE FOLDER NAMED 'Contour'
  cv2.imwrite('/content/gdrive/MyDrive/CA_3/Contour/image'+str(img_number)+'.jpg', img)
  img_number +=1 

  # APPLYING RECTANGULAR BOUNDING BOXES ON THE EROSION IMAGES
  boxes = pytesseract.image_to_boxes(img)
  for b in boxes.splitlines():
    b = b.split(' ')
    img = cv2.rectangle(img, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (255, 0, 0), 2) #OpenCV is in Blue, Green, Red format.
  
  # DISPLAYING THE IMAGES
  #cv2_imshow(img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()

```
